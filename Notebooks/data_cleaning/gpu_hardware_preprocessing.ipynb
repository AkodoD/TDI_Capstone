```python
import pandas as pd
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
link_list = pd.read_json("../../data/raw_data/gpu_urls.json")
raw_list = pd.read_json("../../data/raw_data/final_gpu_list.json")
new_list = link_list.join(raw_list)
```

```python
new_list
```

```python
new_list = new_list[['name','brand','Release Date', 'Generation','Texture Rate','Memory Size','DirectX', 'OpenGL', 'OpenCL']].dropna()
```

```python
from dateutil.parser import parse
def gpu_date_parser(df):
    """cleans the gpu dataframe removing entries without dates and converting the date column to date format"""
    df['Release Date'] = df[df['Release Date'].isin(["Unknown","Never Released"]) == False]['Release Date']
    df = df.dropna().copy()
    df['Release Date'] = df["Release Date"].apply(parse, fuzzy=True)
    return df

def prefixed_number_to_int(cell):
    """used to convert texture rate with letter (100M) to int"""
    val,mod = cell.split()
    mod_val = {"G":1,"M":1/1000}[mod[0]]
    return pd.to_numeric(val)*mod_val
```

```python
new_list.columns
```

```python
df = gpu_date_parser(new_list)
df = df.set_index('Release Date')
df.columns = ["name",'brand','generation',"texture_rate",'memory_size','directx','opengl','opencl']
```

```python
df['texture_rate'] = df['texture_rate'].apply(prefixed_number_to_int)
```

```python
df.loc[(df.memory_size == 'System Shared'),'memory_size'] = '0 MB'
df['memory_size'] = df['memory_size'].apply(prefixed_number_to_int)
```

```python
df.to_pickle("../../data/preprocessed_data/final_gpu_df.pkl")
```
